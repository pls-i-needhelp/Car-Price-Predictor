```python
import numpy as np
import pandas as pd
carData = pd.read_csv('car_dataset.csv')
names = ["Land Rover", "Swift Dzire", "Wagon R", "Vitara Brezza", "Alto 800", "Alto K10", "Zen Estilo", "KUV 100", "TUV 300", "Renault Logan", "Ssangyong Rexton", "Innova Crysta", "Etios Liva", "Corolla Altis", "B Class", "New C-Class", "3 Series", "5 Series", "7 Series", "X4 M", "Indica Vista", "New Safari", "Santro Xing", "Grand i10", "Elite i20", "Figo Aspire", "Fiesta Classic", "Grande Punto", "Tavera Neo", "Sail Hatchback", "GO Plus"]

for i in carData["name"]:
    for j in names:
        if j in i:
            carData["name"] = carData["name"].str.replace(j, j.replace(" ", "-"))

carData.insert(0, "brand", carData["name"].apply(lambda x : x.split(" ")[0]))
carData.insert(1, "model", carData["name"].apply(lambda x : x.split(" ")[1]))

# Create a variant feature by parsing name
carData.insert(2, "variant", carData["name"].apply(lambda x: x.split(" ")[2:]).apply(lambda x: " ".join(x)))
carData.head()
```

```python
carData = carData.iloc[:, :3]
```

```python
carData.head()
```

```python
import csv


carModelAccToBrand = {}
carVariantAccToModel = {}
company = carData['brand'].unique().tolist()

for _, row in carData.iterrows():
    brand = row['brand']
    model = row['model']
    variant = row['variant']
    
    if brand not in carVariantAccToModel:
        carVariantAccToModel[brand] = []
    if model not in carModelAccToBrand[brand]:
        carModelAccToBrand[brand].append(model)

    if model not in carVariantAccToModel:
        carVariantAccToModel[model] = []
    if variant not in carVariantAccToModel[model]:
        carVariantAccToModel[model].append(variant)

print(carModelAccToBrand)
print(carVariantAccToModel)

```

```python
output_file = 'BrandModel.xlsx'
max_length = max(len(arr) for arr in carModelAccToBrand.values())
carModelAccToBrand = {key: arr + [np.nan] * (max_length - len(arr)) for key, arr in carModelAccToBrand.items()}
# Get the header and data from the dictionary
df1 =  pd.DataFrame(carModelAccToBrand)
df1.to_excel(output_file, index=False)
```

```python
output_file = 'ModelVariant.xlsx'
max_length = max(len(arr) for arr in carVariantAccToModel.values())
carVariantAccToModel = {key: arr + [np.nan] * (max_length - len(arr)) for key, arr in carVariantAccToModel.items()}
# Get the header and data from the dictionary
df1 =  pd.DataFrame(carVariantAccToModel)
df1.to_excel(output_file, index=False)
```
